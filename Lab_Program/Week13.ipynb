```python
# Text Analytics and Recommendation System
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')
nltk.download('stopwords')

text = "Data Science is an exciting field. It uses Machine Learning and Artificial Intelligence."
tokens = word_tokenize(text.lower())
stop_words = set(stopwords.words('english'))

filtered_tokens = [
    word for word in tokens
    if word.isalnum() and word not in stop_words
]

print("Original Text:")
print(text)

print("\nTokens:")
print(tokens)

print("\nAfter Stop-word Removal:")
print(filtered_tokens)

documents = [
    "Machine learning for data science",
    "Deep learning and artificial intelligence",
    "Data science with Python",
    "Cooking recipes and food"
]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

similarity_matrix = cosine_similarity(tfidf_matrix)

print("\nCosine Similarity Matrix:")
print(similarity_matrix)

query = "Python for data science"

query_vector = vectorizer.transform([query])

scores = cosine_similarity(query_vector, tfidf_matrix)

recommended_index = scores.argmax()

print("\nQuery:")
print(query)

print("\nRecommended Document:")
print(documents[recommended_index])
```
